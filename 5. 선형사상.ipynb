# 5.1 가우스 조르단 소거법 파이썬 실습

```python
### deepcopy를 구현하기 위한 영행렬 생성 함수
def zero(n, m):
    """영행렬 생성"""
    Z = [] # result matrix
    
    for i in range(0, n):
        row = []
        for j in range(0, m):
            row.append(0)
        Z.append(row)
    
    return Z

zero(3,3)

### 가우스-조르단 소거법을 구현하기 위한 deepcopy 함수
def deepcopy(A):
    """깊은 복사(deepcopy) 구현"""
    if type(A[0]) == list:
        n = len(A)
        m = len(A[0])
        res = zero(n, m) # result matrix
        
        for i in range(0, n):
            for j in range(0, m):
                res[i][j] = A[i][j]
            
        return res
    else:
        n = len(A)
        res = [] #result vector
        for i in range(0, n):
            res.append(A[i])
        return res
    
### 가우스-조르단 소거법 구현 함수
def solve(A, b):
    """연립 방정식 풀기"""
    X = deepcopy(A) # result X
    sol = deepcopy(b) # result sol
    n = len(X)
    
    for i in range(0, n):
        print('i번째 실행 시작')
        x_row = X[i]
        y_val = sol[i]
        
        if x_row[i]!=0:
            temp = 1 / x_row[i]
        else:
            temp = 0
            
        x_row = [element * temp for element in x_row]
        y_val = y_val*temp
        
        X[i] = x_row
        sol[i] = y_val
        
        print(x_row)
        print(y_val)
        print('행 나누기 완료')
        
        for j in range(0, n):
            if i==j:
                continue # 동일한 행에 대해서는 반복문을 수행하지 않고 건너뛴다
                
            print('j번째 실행 시작')
            x_next = X[j]
            y_next = sol[j]
            
            x_temp = [element * -x_next[i] for element in x_row]
            y_temp = y_val * (-x_next[i])
            
            for k in range(0, len(x_row)):
                x_next[k] = x_temp[k] + x_next[k]
            y_next = y_temp + y_next
            
            X[j] = x_next
            sol[j] = y_next
            
            print(X)
            print(sol)
            
            print('j번째 실행 완료')
        print('i번째 실행 완료')
    
    return sol

### 제대로 만들어졌는지 test
A = [[3,1,2],[2,6,-1],[4,0,-1]]
b = [5,1,3]

sol = solve(A, b)
print(sol)
        
```

# 5.2 가우스-조르단 소거법 넘파이 실습

```python
import numpy as np

X = np.array([[3,1,2],[2,6,-1],[4,0,-1]])
y = np.array([[5],[1],[3]])

sol = np.linalg.solve(X, y)

print(sol)
```
